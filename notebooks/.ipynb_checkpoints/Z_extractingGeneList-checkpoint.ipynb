```r
suppressPackageStartupMessages(library(clusterProfiler))
suppressPackageStartupMessages(library(org.Hs.eg.db))
```

```r
data(geneList, package="DOSE")
gene <- names(geneList)[abs(geneList) > 2]
```

```r
gene.df <- bitr(gene, fromType = "ENTREZID",
                toType = c("ENSEMBL", "SYMBOL"),
                OrgDb = org.Hs.eg.db)
head(gene.df)
```

```r
nrow(gene.df)
length(gene)
length(unique(gene.df[,1]))
```

```r
length(make.names(gene.df[,1],unique=TRUE))
df2=cbind(gene.df,make.names(gene.df[,1],unique=TRUE))
```

```r
#df2=gene.df2[order(gene.df2[,4],decreasing = FALSE),]
```

```r
df_u=df2[-grep ('\\.',df2[,4]),]
nrow(df_u)
```
