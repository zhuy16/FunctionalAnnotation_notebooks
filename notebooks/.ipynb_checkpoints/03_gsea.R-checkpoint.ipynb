# 3.3 Cell Marker, helpful for annotation of cell types when needed


vroom is installed on my R/3.5.2 but not on R/3.6.1
cell type markers for mouse and human, and single cells are summarized here:
http://bio-bigdata.hrbmu.edu.cn/CellMarker/
database can be downloaded here, 
http://bio-bigdata.hrbmu.edu.cn/CellMarker/download/

```r
packageVersion("vroom")
```

```r
library(vroom)
cell_markers <- vroom::vroom('http://bio-bigdata.hrbmu.edu.cn/CellMarker/download/Human_cell_markers.txt') %>%
  tidyr::unite("cellMarker", tissueType, cancerType, cellName, sep=", ") %>% 
  dplyr::select(cellMarker, geneID) %>%
  dplyr::mutate(geneID = strsplit(geneID, ', '))
```

```r
head(cell_markers)
nrow(cell_markers)
```

```r
y <- enricher(gene, TERM2GENE=cell_markers, minGSSize=1)
DT::datatable(as.data.frame(y))
```

```r

y <- setReadable(y, org.Hs.eg.db, keyType = "ENTREZID")

head(y)
```
